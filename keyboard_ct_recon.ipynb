```python
import numpy as np
import matplotlib.pyplot as plt

from cil.framework import AcquisitionGeometry
from cil.optimisation.functions import LeastSquares, TotalVariation
from cil.plugins.astra import ProjectionOperator
from cil.optimisation.algorithms import FISTA
from cil.utilities.display import show2D
from cil.utilities.jupyter import islicer
from cil.io import TIFFStackReader
```

### Read keyboard PCA datafile for parameters

```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from util.pcaReader import read_pca_file

filepath = 'keyboard_1/keyboard.pca'
params = read_pca_file(filepath)
```

### Parameters of the CT-scan

```python
from util.getParameters import get_parameters

# Reads all the necessary parameters for the reconstruction.
SOD, SDD, ODD, pixel_size, num_pixels, num_projs, max_k = get_parameters(params)
```

### Setting up the Acquistion geometry

```python
from util.setAcquisitionGeometry import set_acquistion_geometry

ag_big = set_acquistion_geometry(
    source_to_oringin_distance=SOD,
    origin_to_detector_distance=ODD,
    pixel_size=pixel_size,
    num_pixels=num_pixels,
    num_projections=num_projs,
    skip=10)
```

### Read the CT-scan data from the .tif files

```python
from util.readCTData import read_ct_data

aq_data = read_ct_data(
    filename='keyboard_1',
    acquistion_geometry=ag_big,
    num_projections=num_projs,
    skip=10)
```

### Convert the transmission data to the absorption data

```python
from util.dataConverter import convert_data

data = convert_data(aq_data, max_k)
```

### Convert fan-beam sinogram to parallel-beam sinogram 

```python
from util.fanToPara import convert_fan_to_parallel_geometry

Psinogram, Ploc, Pangles = convert_fan_to_parallel_geometry(data, 125, SOD)
```

### Reconstruct using the parallel-beam sinogram

```python
from util.fanToPara import reconstruct_parallel_sinogram

recon = reconstruct_parallel_sinogram(Psinogram=Psinogram, Pangles=Pangles, filter_name='hamming', interpolation='linear', lb=0, ub=0.01, cmap='inferno')
```

### Save parallel-beam sinogram and angles for other testing

```python
import h5py

output_file = 'parallel_beam_data.h5'

with h5py.File(output_file, 'w') as f:
    f.create_dataset('sinogram', data=Psinogram)
    f.create_dataset('angles', data=Pangles)
```
