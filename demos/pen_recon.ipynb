Geometry of the setting

```python
SOD = params['FOD'] # Source-Origin-distance
SDD = params['FDD'] # Source-Detector-distance
ODD = SDD - SOD # Origin-Detector-distance

print(f'SOD: {SOD:.3f} mm')
print(f'SDD: {SDD:.3f} mm')
print(f'ODD: {ODD:.3f} mm')

pixel_size = 200 / 1000
print(f'Detector pixel size: {pixel_size:.3f} mm')

num_pixels = (params['DimX'], params['DimY']) # Width and height of the projection image
num_pixels = (num_pixels[0], num_pixels[1])
print(f'The original size of the projection image: {params['DimX']} x {params['DimY']}, set to {num_pixels[0]} x {num_pixels[1]}')

# Projections
num_projs = params['NumberImages']
print(f'{num_projs} projection angles used')
```

Total Variation (TV) reconstruction algorithm


The goal is to obtain the following: $\hat{x}=\argmin_x \{f(x)+g(x)\}$ where $f$ is a least squares objective function  $\|\boldsymbol{d}-A\boldsymbol{x}|^2_2$
and $g$ is a total variation function $\alpha TV(x)$, where $\alpha$ is a regularization constant. $\hat{x}$ is the reconstructed image, that is close to the true case.

```python
data.reorder('astra')

# Acquisition geometry
ag = data.geometry

A = ProjectionOperator(ig, ag, 'gpu')
f = LeastSquares(A, data)
h = TotalVariation()

# alpha = 1e-5 # Regulazation constant
alpha = 1e-5

tv_fista = FISTA(initial=ig.allocate(0), f=f, g=alpha*h)
```

```python
tv_fista.run(20)
```

```python
x_tv = tv_fista.get_output()
show2D(x_tv, size=(20, 10), fix_range=(0,ub), cmap='inferno', title=f'TV-regularized with {alpha = }', num_cols=3, slice_list=[(0, 150), (1, 170), (2, 95)]);
```

```python
islicer(x_tv, minmax=(0,ub), direction=2, cmap='inferno')
```

```python
from cil.plugins.ccpi_regularisation.functions import FGP_TV

data.reorder('astra')

# Acquisition geometry
ag = data.geometry

A = ProjectionOperator(ig, ag, 'gpu')
f = LeastSquares(A, data)
# h = TotalVariation()
h = FGP_TV(device='gpu', nonnegativity=True)

alpha = 1e-5 # Regulazation constant

tv_fista = FISTA(initial=ig.allocate(0), f=f, g=alpha*h)
```

```python
tv_fista.run(20)
```

```python
x_tv = tv_fista.get_output()
show2D(x_tv, size=(20, 10), fix_range=(0,ub), cmap='gray', title=f'TV-regularized with {alpha = }', num_cols=3, slice_list=[(0, 150), (1, 170), (2, 95)]);
```

```python
islicer(x_tv, minmax=(0,ub), direction=2, cmap='inferno')
```

### PGHD reconstruction algorithm using TV

Without non-negativity constraint

```python
from cil.optimisation.operators import BlockOperator, GradientOperator
from cil.optimisation.algorithms import PDHG
from cil.optimisation.functions import L2NormSquared, BlockFunction, MixedL21Norm, IndicatorBox

# Regularization constant
alpha = 1e-5
D = GradientOperator(ig)

F = BlockFunction(L2NormSquared(b=data), alpha*MixedL21Norm())
K = BlockOperator(A, D)
G = IndicatorBox()

pdhg = PDHG(f=F, g=G, operator=K, update_objective_interval=10, check_convergence=False)
```

```python
pdhg.run(500, verbose=1)
```

```python
show2D(pdhg.solution,
    cmap='inferno',
    fix_range=(0, ub),title=f'PDHG reconstructions with {alpha = }.',
    slice_list=[(0, 150), (1, 170), (2, 95)], num_cols=3);
```

```python
islicer(pdhg.solution, minmax=(0,ub), direction=2, cmap='inferno')
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.loglog(pdhg.iterations[1:], pdhg.objective[1:])
plt.loglog(pdhg.iterations[1:], pdhg.dual_objective[1:])
plt.loglog(pdhg.iterations[1:], pdhg.primal_dual_gap[1:])
# plt.ylim(1e6, 1e7)
plt.legend(['PDHG primal', 'PDHG dual', 'PDHG gap'])
plt.grid()
plt.xlabel('Iterations')
plt.ylabel('Objective value')
```

With non-negativity constraint

```python
# Regularization constant
alpha = 1e-5

# Block function for PDHG
f1 = 0.5 * L2NormSquared(b=data)
f2 = alpha * MixedL21Norm()
F = BlockFunction(f1, f2)

# Block operator for PDHG
K = BlockOperator(A, D)

# Function G for PDHG
G = IndicatorBox(lower=0.0)

# PDHG setup
pdhg = PDHG(f=F, g=G, operator=K, update_objective_interval=10, check_convergence=False)

# Run PDHG
pdhg.run(500, verbose=1)
```

```python
show2D(pdhg.solution,
    cmap='inferno',
    fix_range=(0, ub),title=f'PDHG reconstructions with {alpha = }.',
    slice_list=[(0, 150), (1, 170), (2, 95)],
    num_cols=3);
```

Setup for SPDHG algorithm

```python
# Number of subsets
n_subsets = 10

# Partitioned data
pt_data = data.partition(n_subsets, 'sequential')

# List for containing F_i functions
f_subsets = []

# Defining F_i functions
for i in range(n_subsets):

    fi = 0.5*L2NormSquared(b = pt_data[i])
    f_subsets.append(fi)

# Defining block function F
F = BlockFunction(*f_subsets)

ageom_subset = pt_data.geometry
# Projection operator K
K = ProjectionOperator(ig, ageom_subset)

# Defining G function with non-negativity constraint
G = alpha * FGP_TV(nonnegativity=True)
```

```python
from cil.optimisation.algorithms import SPDHG

spdhg = SPDHG(f = F, g = G, operator = K, update_objective_interval = 10)

spdhg.run(50)

spdhg_recon = spdhg.solution
```

```python
show2D(spdhg_recon, cmap='inferno', fix_range=(0, ub),title=f'SPDHG reconstructions with {alpha = }.',
    slice_list=[(0, 150), (1, 170), (2, 96)],
    num_cols=3);
```

```python
# Number of subsets
n_subsets = 10

# Partitioned data
pt_data = data.partition(n_subsets, 'sequential')

# Lists for containing F_i functions and A_i operators
f_subsets = []
A_subsets = []

# Defining F_i functions
for i in range(n_subsets):

    ageom_subset = pt_data[i].geometry
    Ai = ProjectionOperator(ig, ageom_subset)
    A_subsets.append(Ai)
    # normi = Ai.norm() # Normalizing constant
    # A_subsets.append((1/normi) * Ai)
    # # fi = 0.5 * normi**2 * L2NormSquared(b = (1/normi)*pt_data[i])
    fi = 0.5 * L2NormSquared(b = pt_data[i])
    f_subsets.append(fi)

# Defining F{n+1} and A{n+1}
Grad = GradientOperator(A_subsets[0].domain, backend='c', correlation='SpaceChannel')
# normGrad = Grad.norm()
# Grad_renorm = (1/normGrad) * Grad
# f_reg = normGrad * ig.spacing[0] * alpha * MixedL21Norm()
f_reg = ig.spacing[0] * alpha * MixedL21Norm()

# Defining block function F
F = BlockFunction(*f_subsets, f_reg)

# Projection operator K
# K = BlockOperator(*A_subsets, Grad_renorm)
K = BlockOperator(*A_subsets, Grad)

# Defining probabilities
prob = [1 / (2 * n_subsets)] * n_subsets + [1 / 2]

# Defining G function with non-negativity constraint
G = IndicatorBox(lower=0)
```

```python
spdhg_explicit = SPDHG(f=F, g=G, operator=K, update_objective_interval = 2 * n_subsets, prob=prob)

spdhg_explicit.run(5 * 2 * n_subsets)
```

```python
spdhg_explicit_recon = spdhg_explicit.get_output()
show2D([spdhg_recon, spdhg_explicit_recon], fix_range=(0,1), cmap='inferno', title=['Nomal SPDHG', 'Explicit SPDHG'], slice_list=[(2, 95), (2, 95)])
```
